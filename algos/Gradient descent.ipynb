# Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generate Data

```python
m = 100
X = np.random.randn(m,1)
print(X.shape)
y = 3*X + np.random.randn(m,1) + 10
```

```python
fig, ax = plt.subplots(1, figsize=(5,4))
plt_obj = plt.scatter(X, y, s=1)
```

## Gradient Descent

```python
eta = 0.1
epsilon = 0.001
theta = np.random.randn(2,1)
X_bias = np.c_[np.ones(m), X]  # first element in the intercept

norm = np.inf
models = np.array([theta])

iterations = 0
while norm > eta:
    gradient =  (2/m)*(X_bias.T).dot(X_bias.dot(theta)-y)
    norm = np.linalg.norm(gradient)
    theta = theta - eta*gradient
    models = np.append(models, [theta], axis=0)
    iterations += 1
print("Iterations", iterations)
```

```python
fig, ax = plt.subplots(1, figsize=(5,4))
plt_obj = plt.scatter(X, y, s=1)
for model in models:
    y_pred = model[1]*X + model[0]
    plt.plot(X, y_pred, label=f"{model[1]}x + {model[0]}", alpha=0.2)
plt.show()
```
